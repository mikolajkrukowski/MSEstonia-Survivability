# Katastrofa MS Estonia

Analiza zbioru danych zawierającego informacje o 989 pasażerach oraz członkach załogi, którzy znajdowali się na pokładzie promu MS Estonia w czasie katastrofy do której doszło 28 września 1994 roku. Dane zawierają informacje o tym czy dany pasażer przeżył katastrofe i to jest główne kryterium na podstawie którego będę trenował model klasyfikacji.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('estonia-passenger-list.csv')
df
```

```python
df.describe(include='all')
```

Dzięki metodzie describe() można zauważyć, że średni wiek pasażerów to 44,5 lat, a przeżywalność wynosi tylko 13,8%.

```python
survived = df['Survived'].value_counts()
plt.figure(figsize=(5,7))
survived_df=pd.DataFrame({'Przeżył':[survived[1]], 'Nie przeżył':[survived[0]]})
survived_df.plot(kind='bar')
plt.title('Wykres pokazujący ilu ludzi przeżyło katastrofę')
plt.ylabel('Ilość ludzi', fontsize=15)
plt.xlabel('Przeżywalność', fontsize=15)
plt.show()
```

```python
ppc = sns.barplot(x = df['Country'].value_counts().keys(),
                  y = df['Country'].value_counts().values)
ppc.set_xticklabels(ppc.get_xticklabels(), rotation = 90)
plt.title('Ilość ludzi w zależności od kraju')
plt.figure(figsize=(10,5))
plt.show()
```

Możemy zauważyć, że zdecydowana większość pasażerów i załogi pochodzi ze Szwecji i Estoni.

Gdy używamy funkcji transformującej LabelEncoder normalizujemy wszystkie etykiety.
LabelEncoder przypisuje wartości zaczynając od 0 w kolejności alfabetycznej dlatego warto wcześniej wypisać sobie posortowane etykiety.

```python
df = df.drop(columns  = ['PassengerId', 'Firstname', 'Lastname','Country'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
df1=df
le = LabelEncoder()
df1['Sex'] = le.fit_transform(df1['Sex'])
df1['Category'] = le.fit_transform(df1['Category'])
df1
#Country:   0='Belarus' 1='Canada' 2='Denmark' 3='Estonia' 4='Finland' 5='France' 6='Germany' 7='Great Britain' 8='Latvia' 
#           9='Lithuania' 10='Morocco' 11='Netherlands' 12='Nigeria' 13='Norway' 14='Russia' 15='Sweden'

#Sex:       0='F' 1='M'

#Category:  0='C' 1='P'

#Survived:  0='Not survived' 1='Survived'
```

```python
X = df.drop(columns  = ['Survived'], axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 72, stratify=y)
```

### Algorytm k najbliższych sąsiadów

n_neighbors = 4

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, f1_score

model = KNeighborsClassifier()
model.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
y_pred=model.predict(X_test)
acc_k=accuracy_score(y_pred, y_test)
print('Acc= ', acc_k)
```

```python
l = np.arange(0.1, 1, 0.1)
l = l.round(1)
for i in l:
    y_pred_2 = (model.predict_proba(X_test)[:,1] >= i)
    acc = accuracy_score(y_test, y_pred_2).round(3)
    recall = recall_score(y_test, y_pred_2).round(3)
    f1 = f1_score(y_test, y_pred_2).round(3)
    p = precision_score(y_test, y_pred_2).round(3)
    print('Treshold: ' + str(i) + '\tAccuracy: ' + str(acc) + '\tRecall: ' + str(recall) + '\tF1: ' + str(f1) + '\tPrecision: ' + str(p))
```

```python
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
plot_confusion_matrix(model, X_test, y_test)
plt.show()
```

```python
from sklearn.metrics import plot_precision_recall_curve
plot_precision_recall_curve(model, X_test, y_test)
```
